# Turbine Dataset

```python
df = pd.read_csv("D:\ML\Mini Project\\Turbine_Data.csv")
df.tail()
```

```python
# as 'date column' is named as 'Unnamed: 0'
# duplicate 'Unnamed: 0 col' to 'Date col'
# drop 'Unnamed: 0 col'

df['DateTime'] = df['Unnamed: 0'] 
df.drop('Unnamed: 0', axis=1, inplace=True)

df.drop('WTG', axis=1, inplace=True)
```

```python
df.info()
```

```python
df # NaN values present
```

```python
df.isna().sum()
```

```python
# as ControlBoxTemperature mostly has value 0.0 
df['ControlBoxTemperature'].fillna(0, inplace=True) # replace 0 inplace of NaN for specified column
```

```python
# as AmbientTemperatue mostly has value 27.something
#propagates next valid observation backward (if(AmbientTemperatue[i]) == NaN then set AmbientTemperatue[i]= next non-NaN AmbientTemperatue)
df['AmbientTemperatue'].bfill(inplace=True) 

df['BearingShaftTemperature'].bfill(inplace=True) 

df['Blade1PitchAngle'].bfill(inplace=True) 
```

```python
df.isna().sum()
```

```python
# attribute not needed
df.drop('DateTime', axis=1, inplace=True)
```

# Imputer

```python
# Importing the SimpleImputer class 
from sklearn.impute import SimpleImputer 

# Imputer object using the mean strategy and  
# missing_values type for imputation 
```

```python
#Removing Null Values using IMPUTER MEDIAN METHOD -
imputer = SimpleImputer(strategy='median')
imputer.fit_transform(df)
```

```python
#Removing Null Values using IMPUTER MOST_FREQUENT METHOD -
imputer = SimpleImputer(strategy='most_frequent')
imputer.fit_transform(df)
```

```python
#Removing Null Values using IMPUTER MEAN METHOD -
imputer = SimpleImputer(missing_values = np.nan, strategy ='mean') 

# Fitting the data to the imputer object 
imputed_df = pd.DataFrame(imputer.fit_transform(df))
imputed_df.columns = df.columns
imputed_df.index = df.index

imputed_df.head()
```

```python
imputed_df.isna().sum()
```
