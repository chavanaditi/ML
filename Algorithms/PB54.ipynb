```python
# 2. Add Active Cases column in DataFrame
df['Active cases'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
#df['Active cases'] = ""
df
```

```python
# 3. Write a Python program to get the latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise
df.groupby('Country/Region')['Confirmed','Deaths','Recovered','Active cases'].sum()
                                  
```

```python
# 4. Write a Python program to get the latest number of confirmed deaths and recovered people of Novel Coronavirus (COVID-19) cases Country/Region - Province/State wise
temp1 = df.groupby(['Country/Region','Province/State'])['Confirmed','Deaths','Recovered'].sum()
pd.set_option('display.max_rows', None)
temp1
```

```python
# 5. Write a Python program to get the Chinese province wise cases of confirmed, deaths and recovered cases of Novel Coronavirus (COVID-19)
df_china = df[df['Country/Region']=='China']
df_china = df_china[['Province/State','Confirmed','Deaths','Recovered']]
df_china
```

```python
# 6. Write a Python program to get the top 10 countries data (Last Update, Country/Region, Confirmed, Deaths, Recovered) of Novel Coronavirus (COVID-19)
df.groupby('Country/Region')['Last Update' , 'Confirmed' , 'Deaths' , 'Recovered'].max().sort_values(by='Confirmed',ascending=False)[:10]
```

```python
# 7. Write a Python program to list countries with all cases of Novel Coronavirus (COVID-19) recover
df_recover = df.groupby('Country/Region')['Confirmed','Deaths','Recovered'].sum()
df_recover = df_recover[(df_recover['Confirmed']>0) & (df_recover['Confirmed']==df_recover['Recovered'])]
df_recover
```
