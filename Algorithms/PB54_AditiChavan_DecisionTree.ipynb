```python
import numpy as np
import pandas as pd
dataset = pd.read_csv('C:\\Users\\Aditi\\Desktop\\ML\\dt.csv')
```

```python
X = dataset.iloc[:,:-1] #feature
feature_col = ['age','income','student','credit_rating']
Y = dataset.iloc[:,4] #class tables

from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()

X = X.apply(LabelEncoder().fit_transform) # value transformation: alpha numeric to numeric

from sklearn.tree import DecisionTreeClassifier

regressor = DecisionTreeClassifier(criterion='entropy',max_depth=5,splitter='best') # criterion='entropy' or criterion='gini', max_depth=5 ot max_depth=none, splitter='best or splitter='random
X

# age<30 : 0, 30..40 : 1, >40 :2 
```

```python
regressor.fit(X,Y) # training of classifier
X_in = np.array([1,1,0,0]) #unseen test data
y_pred = regressor.predict([X_in])
print('Prediction : ',y_pred)
```

```python
# conda install -c anaconda graphviz
# conda install -c conda-forge pydotplus
from sklearn.externals.six import StringIO #rep class labels as str values
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()

export_graphviz(regressor,out_file=dot_data,filled=True,rounded=True,special_characters=True,feature_names=feature_col,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('tree.png')
```
